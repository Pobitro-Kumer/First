```python
#Task 2.2: Create and Fix Missing Value
# Artificially create missing data for practice
import numpy as np

# Your tasks:
# 1. Make a copy of the dataset: df_practice = df.copy()
# 2. Randomly set 50 math scores to np.nan
# 3. Randomly set 30 writing scores to np.nan





# Steps to follow:
# Step 1: Create copy and show original missing values
df_practice = df.copy()
print("Original missing values:", df_practice.isnull().sum())

# Step 2: Create missing data
np.random.seed(42)  # For reproducible results
missing_math_idx = np.random.choice(df_practice.index, 50, replace=False)
missing_writing_idx = np.random.choice(df_practice.index, 30, replace=False)

df_practice.loc[missing_math_idx, 'math score'] = np.nan
df_practice.loc[missing_writing_idx, 'writing score'] = np.nan

# 4. Fill missing math scores with the MEDIAN math score
median_math_score = df_practice['math score'].median()
df_practice['math score'] = df_practice['math score'].fillna(median_math_score)


# 5. Fill missing writing scores with the MEAN writing score
mean_writing_score = df_practice['writing score'].mean()
df_practice['writing score'] = df_practice['writing score'].fillna(mean_writing_score)

# 6. Verify no missing values remain
missing_values_exist = df_practice.isnull().sum().sum()
print(f"Total missing values after filling: {missing_values_exist}")
# If missing_values_exist is 0, no missing values remain.

# Continue the task...
# Your code here:
```

```python
#Task 3.2: Basic Groupby Analysis
# Comprehensive groupby analysis
# Your tasks:

# 1. Group by 'gender' and calculate:

#    - Average scores in all subjects

#    - Count of students
gender_analysis = df.groupby('gender').agg(
    avg_math_score=('math score', 'mean'),
    avg_reading_score=('reading score', 'mean'),
    avg_writing_score=('writing score', 'mean'),
    student_count=('gender', 'count')
)
print("1. Gender Analysis:\n", gender_analysis)


# 2. Group by 'test preparation course' and show:
#    - Mean scores for each subject
test_prep_analysis = df.groupby('test preparation course')[['math score', 'reading score', 'writing score']].mean()
print("\n2. Test Preparation Course Analysis:\n", test_prep_analysis)
#    - Show which subject benefits most from test prep
score_diff = test_prep_analysis.loc['completed'] - test_prep_analysis.loc['none']
most_benefiting_subject = score_diff.idxmax()

print(f"\nSubject benefiting most from test prep: {most_benefiting_subject}")


# 3. Group by 'parental level of education' and calculate:
#    - Average math score for each education level
#    - Rank education levels by math performance
parental_education_analysis = df.groupby('parental level of education')['math score'].mean().reset_index()
parental_education_analysis = parental_education_analysis.rename(columns={'math score': 'avg_math_score'})
parental_education_analysis['rank'] = parental_education_analysis['avg_math_score'].rank(method='dense', ascending=False)
parental_education_analysis = parental_education_analysis.sort_values('rank')
print("\n3. Parental Education Analysis:\n", parental_education_analysis)

# 4. Group by 'lunch' type and show:
#    - Average scores for each subject
#    - Count of students in each lunch category
lunch_analysis = df.groupby('lunch').agg(
    avg_math_score=('math score', 'mean'),
    avg_reading_score=('reading score', 'mean'),
    avg_writing_score=('writing score', 'mean'),
    student_count=('lunch', 'count')
)
print("\n4. Lunch Type Analysis:\n", lunch_analysis)

# Your code here:
```

```python
#Task 3.3: Use your data analysis skills to answer these easy questions:

# 1. Who performs better in math - males or females?

#    - Calculate average math score by gender
#    - Show the difference between male and female average
average_math_by_gender = df.groupby('gender')['math score'].mean()
print("Average math Score by gender:\n", average_math_by_gender)
print("Difference between male and female average:", average_math_by_gender.loc['male'] - average_math_by_gender.loc['female'])

# 2. Which subject do students perform best in overall?
#    - Calculate the overall average for math, reading, and writing
#    - Rank the subjects from highest to lowest average score
overall_avg_scores = df[['math score', 'reading score', 'writing score']].mean()
print("\nOverall average scores by subject:\n", overall_avg_scores)
ranked_subjects = overall_avg_scores.sort_values(ascending=False)
print("\nSubjects ranked by average score:\n", ranked_subjects)

# 3. What's the impact of parental education?
#    - Find the average total score for each parental education level
#    - Identify which education level leads to highest student performance
df['total score'] = df['math score'] + df['reading score'] + df['writing score']
average_total_score_by_parental_education = df.groupby('parental level of education')['total score'].mean().sort_values(ascending=False)
print("\nAverage total score by parental education level:\n", average_total_score_by_parental_education)
print(f"\nParental education level leading to highest student performance: {average_total_score_by_parental_education.index[0]}")
```
